# Problem 1 — Mobile Recharge Validation System

Validate a mobile recharge request.

Rules:
- Amount must be ≥ 50
- Valid plans: 199, 299, 399, 599
- Allow retry if invalid

Real-world use: Telecom recharge apps

```python
def validate_recharge():
    valid_plans = [199, 299, 399, 599]

    while True:
        amount = int(input("Enter recharge amount: "))

        if amount < 50:
            print("Amount must be at least ₹50")
            continue

        if amount not in valid_plans:
            print("Invalid plan. Try again.")
            continue

        print("Recharge successful ✅")
        break

validate_recharge()
```

# Problem 2 — Inventory Reorder Alert System

Check stock levels and print:
Reorder Alert if stock < 15
Stock OK otherwise

```python
def check_inventory(products):
    for product, qty in products.items():
        if qty < 15:
            print(f"{product}: Reorder Alert")
        else:
            print(f"{product}: Stock OK")

products = {
    "Apples": 10,
    "Bananas": 25,
    "Oranges": 12
}

check_inventory(products)
```

# Problem 3 — Student Result Processing System

Calculate average marks.
Print Pass if average ≥ 50 else Fail.

```python
def process_result(marks):
    total = 0

    for m in marks:
        total += m

    average = total / len(marks)

    print("Average:", average)

    if average >= 50:
        print("Result: Pass")
    else:
        print("Result: Fail")

marks = [60, 45, 70, 50]
process_result(marks)
```

# Problem 4 — Cab Fare Estimator with Retry

Base fare = 50
₹12 per km
Peak hour = +25%
Allow retry if user wants

```python
def calculate_fare(distance, peak):
    fare = 50 + (distance * 12)

    if peak.lower() == "yes":
        fare *= 1.25

    return fare


while True:
    distance = float(input("Enter distance in km: "))
    peak = input("Peak hour? (yes/no): ")

    fare = calculate_fare(distance, peak)
    print("Total Fare:", fare)

    retry = input("Calculate again? (yes/no): ")
    if retry.lower() != "yes":
        break
```

# Problem 5 — Employee Attendance Eligibility

Calculate attendance percentage.
Eligible if ≥ 75%

```python
def check_attendance(attendance):
    present = 0

    # Loop to count present days
    for day in attendance:
        if day == "P":
            present += 1

    percentage = (present / len(attendance)) * 100

    return percentage


attendance = ["P","A","P","P","A","P","P","P"]

percentage = check_attendance(attendance)

print("Attendance %:", percentage)

if percentage >= 75:
    print("Eligible")
else:
    print("Not Eligible")
```

# Problem 6 — Password Strength Checker

Rules:
- Minimum length 8
- At least one digit
- At least one special character (@#$)

```python
# Problem 6 — Password Strength Checker

def check_password(password):
    has_digit = False
    has_special = False
    special_chars = "@#$"

    if len(password) < 8:
        return "Weak password ❌"

    for ch in password:
        if ch.isdigit():
            has_digit = True
        if ch in special_chars:
            has_special = True

    if has_digit and has_special:
        return "Strong password ✅"
    else:
        return "Weak password ❌"


# Taking input from user
password = input("Enter your password: ")

result = check_password(password)
print(result)
```
