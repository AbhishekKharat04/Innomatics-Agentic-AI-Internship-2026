## Problem 1: Unique Words in a Sentence
This program finds all unique words and their count.


```python
def unique_words(sentence):
    words = sentence.split()
    unique = set(words)

    print("Unique words count:", len(unique))
    print("Unique words:", unique)


sentence = "python is easy and python is powerful"
unique_words(sentence)
```

## Problem 2: Highest Salary from Employee Data


```python
def highest_salary(employees):
    name = max(employees, key=employees.get)
    print("Highest Salary:", name, "-", employees[name])


employees = {
    "Ravi": 75000,
    "Anita": 68000,
    "Kiran": 72000
}

highest_salary(employees)
```

## Problem 3: Maximum and Minimum Values


```python
def find_max_min(numbers):
    maximum = max(numbers)
    minimum = min(numbers)

    print("List:", numbers)
    print("Max:", maximum)
    print("Min:", minimum)


numbers = [45, 22, 89, 10, 66]
find_max_min(numbers)
```

## 4. Count Products Above Price Threshold
This program counts how many products have a price greater than 1000.



```python
def count_expensive(prices):
    count = 0
    for price in prices:
        if price > 1000:
            count += 1
    print("Products above 1000:", count)


prices = [450, 1200, 899, 1500, 300]
count_expensive(prices)
```

## 5. Calculate Attendance Percentage
This program calculates attendance percentage based on the number of present days.


```python
def attendance_percentage(attendance):
    present = attendance.count("P")
    percentage = (present / len(attendance)) * 100
    print("Attendance Percentage:", percentage)


attendance = ["P", "P", "A", "P", "P"]
attendance_percentage(attendance)
```

## 6. Remove Duplicate Phone Numbers
This program removes duplicate phone numbers and displays only unique values.


```python
def unique_numbers(numbers):
    unique = set(numbers)
    print("Unique phone numbers:", unique)


phone_numbers = [9876543210, 9123456789, 9876543210]
unique_numbers(phone_numbers)
```

## 7. Count Character Frequency
This program counts how many times each character appears in a given string.


```python
def char_frequency(text):
    freq = {}

    for char in text:
        freq[char] = freq.get(char, 0) + 1

    print(freq)


char_frequency("python")
```

## 8. Convert List to Tuple

## 9. Check if a Key Exists in a Dictionary
This program checks whether a specified key exists in a dictionary.


```python
def check_key(dictionary, key):
    if key in dictionary:
        print("Key exists")
    else:
        print("Key not found")


data = {"name": "Ravi", "age": 25}
check_key(data, "name")
```

## 10. Calculate Average Marks
This program calculates the average marks from a list of scores.


```python
def average_marks(marks):
    avg = sum(marks) / len(marks)
    print("Average Marks:", avg)


marks = [70, 80, 65, 75]
average_marks(marks)

```
