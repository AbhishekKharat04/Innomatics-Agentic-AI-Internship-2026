## 2. Pass / Fail Analyzer
This program analyzes student marks and counts the number of pass and fail students.
Passing condition: marks >= 50


```python
def analyze_results(marks):
    pass_count = 0
    fail_count = 0

    for mark in marks:
        if mark >= 50:
            pass_count += 1
        else:
            fail_count += 1

    print("\n----- Result Summary -----")
    print(f"Total Students: {len(marks)}")
    print(f"Passed Students: {pass_count}")
    print(f"Failed Students: {fail_count}")



# Given data
marks = [45, 78, 90, 33, 60]

# Function call
analyze_results(marks)

```

## 3. Simple Data Cleaner
This program removes extra spaces and converts names to lowercase.


```python
names = [" Alice ", "bob", " CHARLIE "]

cleaned_names = []

for name in names:
    cleaned_name = name.strip().lower()
    cleaned_names.append(cleaned_name)

print("Cleaned Names:", cleaned_names)
```

## 4. Message Length Analyzer
This program prints the length of each message and flags messages longer than 10 characters.


```python
def analyze_messages(messages):
    for message in messages:
        length = len(message)
        print(f"Message: '{message}' | Length: {length}")

        if length > 10:
            print("⚠️ This message is longer than 10 characters")


# Given data
messages = ["Hi", "Welcome to the platform", "OK"]

# Function call
analyze_messages(messages)
```

## 5. Error Message Detector
This program counts the number of "ERROR" entries in logs.


```python
logs = ["INFO", "ERROR", "WARNING", "ERROR"]

error_count = 0

for log in logs:
    if log == "ERROR":
        error_count += 1

print("Total ERROR messages:", error_count)
```
